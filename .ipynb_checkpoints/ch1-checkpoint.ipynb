```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()# Bunch class object
```

```python
print(iris_dataset.keys())
```

```python
#in DESCR, there is simple characteristics for datasets
print(iris_dataset['DESCR'][:250] + "\n...")
```

```python
print(iris_dataset['data'][:5])
```

```python
print(iris_dataset['target'])
```

```python
from sklearn.model_selection import train_test_split
# random_state=0 -> ëžœë¤. 75%, 25%ë¡œ split í•˜ë©´ ë’· ë¶€ë¶„ì€ í´ëž˜ìŠ¤ 2 ë§Œ ë“¤ì–´ê°€ê¸° ë•Œë¬¸ì—.
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
# ì‚°ì ë„ ê·¸ë¦¬ê¸° ìœ„í•´ numpy arrayë¥¼ pandas dataframeìœ¼ë¡œ ë³€ê²½ í›„ ì¶œë ¥
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10,10), marker='o',
                          hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
# k-NN (k-Nearest Neighbors)
# sklearnì˜ ëª¨ë“  ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ì€ Estimatorë¼ëŠ” íŒŒì´ì¬ í´ëž˜ìŠ¤ë¡œ êµ¬í˜„ë˜ì–´ ìžˆë‹¤
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# knn ê°ì²´ì˜ fit ë©”ì†Œë“œë¥¼ í†µí•´ ëª¨ë¸ ìƒì„±
knn.fit(X_train, y_train)
```

```python
# ì˜ˆì¸¡í•˜ê¸°
# ìž„ì˜ì˜ ë°ì´í„°ë¥¼ í•˜ë‚˜ ë§Œë“ ë‹¤
# sklearnì€ í•­ìƒ ë°ì´í„°ê°€ 2ì°¨ì› ë°°ì—´ì¼ ê²ƒìœ¼ë¡œ ì˜ˆìƒí•˜ë¯€ë¡œ 2ì°¨ì› ë°°ì—´ë¡œ ë§Œë“ ë‹¤.
X_new = np.array([[5, 2.9, 1, 0.2]])
print(X_new.shape)
```

```python
prediction = knn.predict(X_new)
print(prediction)
print(iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print("prediction for testset", y_pred)
```

```python
print("performance", np.mean(y_pred == y_test))
```

```python
print("performance", knn.score(X_test, y_test))
```

```python
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train, y_train)

print("performance", knn.score(X_test, y_test))
```
